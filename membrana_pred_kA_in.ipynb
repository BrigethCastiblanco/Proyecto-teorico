```python
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
import membrana_potencial_redu as potencial
```

```python
#Constantes

A = potencial.A
B = potencial.B
C = potencial.C

lam_debye = potencial.lam_debye
d = potencial.d
N_pep = potencial.N_pep
e_a = potencial.e_a
e_m = potencial.e_m
k = potencial.k
T = potencial.T
e = potencial.e

B
```

```python
def K_A_in(D_p,pHA,phiA,r):
    KAin = 0.6*10**(-pHA)*4*r*D_p*np.exp(phiA*1e3*0.0389)
    return KAin
```

```python
D_p = 9.3e9     #Constante de difusin de protones
r = 0.5     #Radio del canal de protones
pHA = 6.6         #pH del canal de entrada

for m in A:
    print(K_A_in(D_p,pHA,B[m],r))
```

```python
#K_in desde la parte del periplasma (sitio casi extracelular)
phiA = []
kinA =[]
for m in B:
    phi_A = B[m]
    Dphi_A = B[0] - B[m]
    phiA.append(phi_A*1e3)
    K_in = K_A_in(D_p,pHA,Dphi_A,r)
    kinA.append(K_in)
    print(Dphi_A,phi_A)

#grfica respecto al poetencial sobre la superficie z=d
plt.plot(phiA,kinA)

#plt.legend()
plt.xlabel(r"$\phi_s$ $(mV)$")
plt.ylabel(r"$k_{in}$ ($s^{-1}$)")
plt.grid()
plt.show()
    
```

```python
#comprobacin comportamiento exponencial
plt.semilogy(phiA,kinA)
plt.xlabel("")
plt.ylabel("")
plt.grid()
plt.show()
```
